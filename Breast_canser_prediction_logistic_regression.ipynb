```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn.datasets
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression

from sklearn import metrics
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score, precision_score,recall_score,f1_score
```

```python
breast_cancer_dataset=sklearn.datasets.load_breast_cancer()
```

```python
bcd = pd.DataFrame(breast_cancer_dataset.data, columns=breast_cancer_dataset.feature_names)
```

```python
X=bcd.drop(columns='lebel',axis=1)
y=bcd["lebel"]
```

```python
print(X)
```

```python
x_train,x_test,y_train,y_test=train_test_split(X,y,test_size=0.25,random_state=3)
```

```python
print(x_train.shape,x_test.shape)
```

```python
model=LogisticRegression()
```

```python
model.fit(x_train,y_train)
```

```python
x_train_pred=model.predict(x_train)
```

```python
accuracy=accuracy_score(y_train,x_train_pred)
```

```python
accuracy
```

```python
x_test_pred=model.predict(x_test)
```

```python
testaccuracy=accuracy_score(y_test,x_test_pred)
```

```python
testaccuracy
```

```python
cm=confusion_matrix(y_test,x_test_pred)


sns.heatmap(cm,annot=True,fmt='g',xticklabels=['malignant','benign'],yticklabels=['malignant','benign'])plt.ylabel("prediction")
plt.xlabel("actual")
plt.title("Confusion Matrix")
plt.show()
```

```python
f1 = f1_score(y_test, x_test_pred)
precision = precision_score(y_test, x_test_pred)
recall = recall_score(y_test, x_test_pred)
print("F1 Score:", f1)
print("Precision Score:", precision)
print("Recall Score:", recall)
```

```python
from sklearn import metrics
from sklearn.metrics import roc_curve, auc

# Calculate ROC curve and AUC
fpr, tpr, thresholds = roc_curve(y_test, x_test_pred)
roc_auc = auc(fpr, tpr)

# Plot ROC curve
plt.figure()
lw = 2
plt.plot(fpr, tpr, color='darkorange',linestyle='--',
         lw=lw, label='ROC curve (area = %0.2f)' % roc_auc)
plt.plot([0, 1], [0, 1], color='navy', lw=lw, linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Breast canser prediction')
plt.legend(loc="lower right")
plt.show()
```
